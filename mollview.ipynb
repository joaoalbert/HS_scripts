```python
import healpy as hp
import numpy as np
import matplotlib as plt
```

```python
nside = 128
npix = hp.nside2npix(nside)
vec = np.arange(npix)

hp.mollview(vec, cmap = 'jet')
hp.mollview(vec, cmap = 'cool')
hp.mollview(vec, cmap = 'winter')
```

```python
hp.cartview(vec, cmap = 'winter')
```

```python
vec = np.zeros(npix)
for i in range(360): 
    pix = hp.ang2pix(nside,theta = i, phi = -22, lonlat = True)
    vec[pix] = 50000
    
hp.oth(vec, rot = [0,-90], xsize = 100)
```
